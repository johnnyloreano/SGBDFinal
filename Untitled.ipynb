## Useful imports

```python
import pandas as pds
import matplotlib.pyplot as plt
```

# Preparing the Dataset
## Creating all necessaries columns name

```python
column_names_quantitatives = ["Elevation","Aspect","Slope","HorDistToHydro","VertDistToHydro","HorDistRoad","Hillshade09","Hillshade12","Hillshade15","HorDistFire"]
column_names_categoric = ["WA_RWA","WA_NWA","WA_CPWA","WA_CLPWA"]
column_names_soil_type = ["Soil_Type" + str(x) for x in range(40)]
column_names_cover_type = ["Cover_Type"]
all_column = column_names_quantitatives + column_names_categoric + column_names_soil_type + column_names_cover_type
```

## Reading CSV

```python
dataset = pds.read_csv('covtype.data/covtype.data', names = all_column, nrows = 300)
```
